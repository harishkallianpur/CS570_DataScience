##Network Analysis Homework
<p style="color:blue">
Import networkx and pyplot for Graph analysis and plotting
</p>

```python
import networkx as nx
import matplotlib.pyplot as plt
%matplotlib inline
```

<p style="color:blue">
Import Political Blogs dataset as a Directed Graph
</p>

```python
M=nx.read_gml("polblogs.gml")
```

<p style="color:blue">
Create a weighted graph from M 
</p>

```python
G = nx.Graph()
for u,v,data in M.edges_iter(data=True):
    if G.has_edge(u,v):
        G[u][v]['weight'] += 1
    else:
        G.add_edge(u, v, weight=1)
```

### 1.  Create a nice looking graph of the political blog network

<p style="color:blue">
Plot weighted graph. Logic copied from Boston-1776 example by Professor Michael Bernico
</p>

```python
plt.figure(figsize=(40,40))
pos=nx.spring_layout(G, iterations=500, scale=5, k=.3)
nx.draw_networkx_labels(G,pos,fontsize=8)
font = {'fontname'   : 'Helvetica',
            'color'      : 'k',
            'fontweight' : 'bold',
            'fontsize'   : 14}
plt.title("Political Blog Network", font)


nx.draw_networkx_nodes(G,pos,node_color='b',alpha=0.4)
nx.draw_networkx_edges(G,pos,alpha=0.4,node_size=0,width=1,edge_color='k')
plt.axis('off')
plt.show()
```

<p style="color:blue">
Compute Graph centrality for G
</p>

```python
centrality=nx.eigenvector_centrality(G)
```

<p style="color:blue">
Build a blog list with tuples containing blog name and eigenvector centrality value
</p>

```python
blog_list = []
for node in centrality:
    blog_list.append((node,centrality[node]))
```

### 2.  What are the 10 most influential blogs?

<p style="color:blue">
Sort the blog list in descending order of eigenvector centrality value. This will give most influential blogs on the top. Show the top 10
</p>

```python
blog_list = sorted(blog_list, key=lambda x: x[1], reverse=True)
```
